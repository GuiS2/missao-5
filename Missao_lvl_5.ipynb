```python
#Gerando os Dados e Salvando-os em um arquivo:
from num2words import num2words
from faker import Faker
from datetime import datetime
fake= Faker('pt_BR')
notas_ext_list= [];
my_file = open('Dados.txt','w+')
for i in range(20):
 nomes= (fake.name())
 notas = (fake.random_int(1,10)) 
 dados = nomes,notas
 notas_extenso = num2words(notas,lang='pt_BR') 
 notas_ext_list.append(notas_extenso)
 notas_str = (' ').join(notas_ext_list)
 print(dados)
```

```python
#Visualizando em Histograma:
import matplotlib.pyplot as plt 
lst_anos = notas_ext_list
num_bins=50; plt.hist(lst_anos,num_bins,density=True,facecolor='purple',alpha=1)
plt.xlabel('Notas')
plt.ylabel('probabilidade')
plt.title('histograma das Notas')
plt.grid(True)
plt.show()
```

```python
#Visualizando em Nuvem De Palavras:
from wordcloud import WordCloud
nuvem_palavras = WordCloud(background_color='White', width=600,height=400).generate(notas_str)
plt.imshow(nuvem_palavras, interpolation='bilinear')
plt.axis("off")
plt.show()
nuvem_palavras.to_file("Nuvem de palavras.png")
```
